#### Utiliser pandas pour  séparer les données afin de les mettre dans plusieurs tables d'une base de donnée que vous allez appeler pubmed (en faisant noramlizant vos données).

```python
import pandas as pd
import numpy as np
import sqlite3

df = pd.read_csv('pubmed.csv')
df.drop('Unnamed: 0', axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
lbe = LabelEncoder()
lbe.fit_transform(df.journal)
```

```python
df['journal_id'] = lbe.fit_transform(df.journal)
```

```python
df2 = df[['pubmed_id','title', 'abstract', 'publication_date','keywords', 'doi', 'last_author', 'journal_id']]
df2.head()
```

```python
df3 = pd.DataFrame({'Journal': lbe.classes_})
df3.insert(0, 'id_journal', range(len(df3)))
df3
```

```python
try:
    # La méthode connect() permet de créer une connexion à la base de donnée
    sqliteConnection = sqlite3.connect('ApiGene.db')
    articles = '''CREATE TABLE Articles(
                        ID INTEGER PRIMARY KEY AUTOINCREMENT,
                        pudmed_id VARCHAR NOT NULL,
                        title VARCHAR NOT NULL,
                        abstract VARCHAR NOT NULL,
                        publication_date date,
                        keywords VARCHAR NOT NULL,
                        doi VARCHAT NOT NULL,
                        last_author VARCHAR NOT NULL,
                        journal_id INTEGER NOT NULL,
                        FOREIGN KEY (journal_id) REFERENCES Journaux (id_journal));'''
    
    journal = '''CREATE TABLE Journaux(
                        ID INTEGER PRIMARY KEY AUTOINCREMENT,
                        journal VARCHAR NOT NULL);'''

    cursor = sqliteConnection.cursor()
    print('Database Created and Successful Connected to SQLite')    
    df2.to_sql("Articles", sqliteConnection, schema=articles, if_exists="replace")
    df3.to_sql("Journaux", sqliteConnection, schema=journal, if_exists="replace")
    sqliteConnection.commit()
    
    print("SQLite table created")
    cursor.close()
    
except sqlite3.Error as error:
    print("Error while connecting to sqlite", error)
finally:
    if(sqliteConnection):
        sqliteConnection.close()
        print("The SQLite connection is closed !")
```

```python
conn = sqlite3.connect('ApiGene.db')
dfArticles = pd.read_sql_query('SELECT * FROM Articles', conn, index_col="index")
dfArticles
```

```python
dfJournal = pd.read_sql_query('SELECT * FROM Journaux', conn, index_col="id_journal")
dfJournal.drop('index', axis=1)
```

```python
joinTables = conn.execute("SELECT title, Journal FROM Articles INNER JOIN Journaux on Articles.journal_id = Journaux.id_journal")
dfJoin = pd.DataFrame(joinTables)
dfJoin
```
