# Mini-projet

Récupérer Titre, abstract, pubmed_id, publication_date, keywords, journal, doi, dernier auteur (le plus a gauche) de différents articles (au moins 1000) scientifiques via l'api pubmed, avec la recherche par thématique en utilisant ""gene""

```python
import requests
import json
from xml.etree import ElementTree
from bs4 import BeautifulSoup
import pandas as pd
from datetime import datetime
```

```python
def get_url(db, q):
    BASE_URL = "https://eutils.ncbi.nlm.nih.gov/entrez/eutils/"
    return BASE_URL  + f"esearch.fcgi?db={db}&term={q}&retmax=1000&sort=relevance&retmode=json"

URL = get_url("pubmed","gene")
```

```python
req = requests.get(URL).content
#req = requests.get(URL).json
results = json.loads(req)
ids = results['esearchresult']['idlist']
```

```python
slicer = 1000//5
get_ids = [",".join(ids[portion:portion+slicer]) for portion in range(0,1000,slicer)]
get_ids
```

```python
def get_infos_from_id(id_list, rettype=None, decode=None):
    slicer = 1000//5
    get_ids = [",".join(id_list[portion:portion+slicer]) for portion in range(0,1000,slicer)]
    if rettype == "xml":
        reqs = []
        for pack_id in get_ids:
            get_url = f"https://eutils.ncbi.nlm.nih.gov/entrez/eutils/efetch.fcgi?db=pubmed&id={pack_id}&rettype=xml"
            req = requests.get(get_url, stream=True)
            if decode:
                req.raw.decode_content = True
                reqs.append(ElementTree.iterparse(req.raw))
            else:
                reqs.append(req.content)
        return reqs
    elif rettype=="text":
        get_url = f"https://eutils.ncbi.nlm.nih.gov/entrez/eutils/efetch.fcgi?db=pubmed&id={get_ids}"
        return requests.get(get_url).text
    else:
        get_url = f"https://eutils.ncbi.nlm.nih.gov/entrez/eutils/efetch.fcgi?db=pubmed&id={get_ids}"
        return requests.get(get_url)
```

```python
list_infos = get_infos_from_id(ids, "xml")
```

```python
library_soup = [BeautifulSoup(pack_infos, 'lxml') for pack_infos in list_infos]
```

```python
library_soup[0].find_all('pubmedarticleset')[0].find_all("author")[-1]
```

```python
datetime.strptime("/".join(library_soup[0].find_all('pubmedarticle')[0].find('pubmedpubdate', {"pubstatus" : "pubmed"}).text.split()), '%Y/%m/%d/%H/%M')
```

```python
pubmed_data = {
    'title' : [],
    'abstract' : [],
    'pubmed_id' : [],
    'publication_date' : [],
    'keywords' : [],
    'journal' : [],
    'doi' : [],
    'last_author' : []
}

for soup in library_soup:
    for article in soup.find_all('pubmedarticle'):
        pubmed_data['title'].append(article.articletitle.text)
        pubmed_data['abstract'].append(article.abstract.text)
        pubmed_data['pubmed_id'].append(int(article.pmid.text))
        pubmed_data['publication_date'].append( datetime.strptime("/".join(article.find('pubmedpubdate', {"pubstatus" : "pubmed"}).text.split()), '%Y/%m/%d/%H/%M') )
        try:
            pubmed_data['keywords'].append(", ".join(article.keywordlist.text.split()))
        except:
            print("No keywords available for this article")
            pubmed_data['keywords'].append("")
        pubmed_data['journal'].append(article.journal.title.text.strip() + ', ' + article.journal.isoabbreviation.text.strip())
        try:
            pubmed_data['doi'].append(article.find('elocationid', {'eidtype': 'doi'}).text)
        except:
            print("Error : couldn't find the DOI")
            pubmed_data['doi'].append("")
        pubmed_data['last_author'].append(article.authorlist.author.lastname.text + ', ' + article.authorlist.author.forename.text)
    

```

```python
pubmed = pd.DataFrame(pubmed_data)
pubmed
```

```python
#Save as CSV
pubmed.to_csv('pubmed.csv')
```

### Faire des requêtes SQL 

1. pour récupérer les abstracts des auteurs ayant au moins publiés deux fois.
2. pour récupérer un classement des journaux scientifiques par rapport aux nombres d'articles publiés
3. pour récupérer le nombre d'articles publiés par an

```python
conn = sqlite3.connect('simplon.db')
c = conn.cursor()
```

```python
# récupérer les abstracts des auteurs ayant au moins publiés deux fois.
c.execute("SELECT abstract FROM pubmed GROUP BY last_author HAVING COUNT(last_author) >1")
```

```python
c.execute("SELECT journal, COUNT(title) FROM pubmed GROUP BY journal ORDER BY COUNT(title) DESC")
```
