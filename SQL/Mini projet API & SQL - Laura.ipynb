### Création de la BDD: 

Utiliser pandas pour séparer les données afin de les mettre dans plusieurs tables d'une base de donnée que vous allez appeler pubmed (en faisant normalisant vos données).


```python
import pandas as pd
```

```python
pubmed_data = pd.read_csv('/Users/Johanna/Documents/SIMPLON DATA IA/IA/ML PROJET/PROJET API SQL/pubmed.csv')
pubmed_data = pubmed_data.drop(['Unnamed: 0'], axis=1)
```

```python
pubmed_data.head()
```

```python
pubmed_data["last_author"] = pubmed_data["last_author"].astype('category')
pubmed_data["journal"] = pubmed_data["journal"].astype('category')

pubmed_data["last_author"].dtypes
```

```python
pubmed_data["auteur_id"] = pubmed_data["last_author"].cat.codes
pubmed_data["journal_id"] = pubmed_data["journal"].cat.codes
pubmed_data.head()
```

```python
pubmed_data['journal'].value_counts()
```

```python
len(pubmed_data['journal'].unique())
```

```python
journal_order = pd.DataFrame(pubmed_data[["journal","journal_id"]])
journal_order = journal_order.sort_values(by=['journal_id'])
journal_unique = journal_order.drop_duplicates(subset=["journal","journal_id"])
journal_unique.shape
```

```python
pubmed_data['last_author'].value_counts()
```

```python
author_unique = list(pubmed_data['last_author'].unique())
auteur_df = pd.DataFrame(author_unique, columns=['auteur'])
auteur_df
```

```python
author_order = pd.DataFrame(pubmed_data[["last_author","auteur_id"]])
author_order = author_order.sort_values(by=['auteur_id'])
author_unique = author_order.drop_duplicates(subset=["last_author","auteur_id"])
author_unique.shape
author_unique
```

```python
pubmed_data['publication_date'].value_counts()
```

```python
# Split column
author_df = pd.DataFrame(author_unique)
author_df
```

```python
publication_df = pd.DataFrame(pubmed_data[['pubmed_id','doi','title','abstract','keywords','publication_date','auteur_id','journal_id']])
publication_df
```

```python
journal_df = pd.DataFrame(journal_unique, columns=["journal","journal_id"])
journal_df
```

```python
import sqlite3

conn = sqlite3.connect('pubmed.db')  # You can create a new database by changing the name within the quotes
c = conn.cursor() # The database will be saved in the location where your 'py' file is saved

# Create table - PUBLICATION
c.execute('''CREATE TABLE IF NOT EXISTS PUBLICATION
             ([pubmed_id] INTEGER PRIMARY KEY,[doi] text, [title] text, [abstract] text, [keywords] text, [publication_date] date, [auteur_id] INTEGER NOT NULL, [journal_id] INTEGER NOT NULL, 
             FOREIGN KEY (auteur_id) REFERENCES AUTEUR (auteur_id),
             FOREIGN KEY (journal_id) REFERENCES JOURNAL (journal_id))''')
          
# Create table - AUTEUR
c.execute('''CREATE TABLE IF NOT EXISTS AUTEUR
             ([auteur_id] INTEGER PRIMARY KEY,[auteur] text)''')
        
# Create table - JOURNAL_TABLE
c.execute('''CREATE TABLE IF NOT EXISTS JOURNAL
             ([journal_id] INTEGER PRIMARY KEY, [journal] text)''')
                 
conn.commit()

# Note that the syntax to create new tables should only be used once in the code (unless you dropped the table/s at the end of the code). 
# The [generated_id] column is used to set an auto-increment ID for each record
# When creating a new table, you can add both the field names as well as the field formats (e.g., Text)
```

```python
publication_df.to_sql('PUBLICATION', con=conn, if_exists='append', index = False)
# Insert the values from the dataframe into the table 'PUBLICATION'

auteur_df.to_sql('AUTEUR', con=conn, if_exists='append', index = False)
journal_df.to_sql('JOURNAL_TABLE', con=conn, if_exists='append', index = False)
```

### Faire des requêtes SQL

- pour récupérer les abstracts des auteurs ayant au moins publiés deux fois.
- pour récupérer un classement des journaux scientifiques par rapport aux nombres d'articles publiés
- pour récupérer le nombre d'articles publiés par an

```python
# https://www.guru99.com/sqlite-query.html
# https://www.sqlitetutorial.net/sqlite-count-function/
```

```python
c.execute('''SELECT DISTINCT abstract
             FROM PUBLICATION''').fetchall()

c.execute('''SELECT auteur_id,
             COUNT(*)
             FROM PUBLICATION
             GROUP BY auteur_id;''').fetchall()

c.execute('''SELECT abstract,
             COUNT(*)
             FROM PUBLICATION
             GROUP BY auteur_id
             HAVING COUNT(*) >= 2;''').fetchall()
```

```python
# pour faire apparaitre tous les abstracts qui sont supperieur à 1
c.execute('''Select abstract from PUBLICATION join (SELECT abstract,
             COUNT(*)
             FROM PUBLICATION
             GROUP BY auteur_id
             HAVING COUNT(*) >= 2;) 
             AS TWO_AUTHORS ON PUBLICATION.auteur_id = TWO_AUTHORS.auteur_id;''').fetchall()
#Shady: SELECT abstract FROM pubmed
#JOIN (SELECT last_author FROM pubmed GROUP BY last_author HAVING COUNT(last_author) > 1)
#AS two_authors ON pubmed.last_author=two_authors.last_author
```

```python
c.execute('''SELECT journal, 
             COUNT(*)
             FROM JOURNAL_TABLE
             INNER JOIN PUBLICATION ON PUBLICATION.journal_id = JOURNAL_TABLE.journal_id
             GROUP BY PUBLICATION.journal_id
             ORDER BY COUNT(*) DESC;''').fetchall()
```

```python
# http://aprogrammersday.blogspot.com/2014/02/group-by-month-or-year-and-sum-in-sqlite.html
# https://kimsereyblog.blogspot.com/2016/06/nice-features-and-tricks-with-sqlite.html

c.execute('''SELECT strftime('%Y', publication_date) as valYear, 
             COUNT(*) as valTotalYear 
             FROM PUBLICATION
             GROUP BY valYear;''').fetchall()
```
