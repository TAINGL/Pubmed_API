### pour récupérer le nombre d'articles publiés par an

```python
# récupérer le nombre d'articles publiés par an SI FORMAT CORRECT
# https://docs.mongodb.com/manual/reference/operator/aggregation/year/
db.pubmed_article.aggregate([{$project:{year:{$year: "$publication_date"}}},{$group:{_id:"$year", total:{$sum:1}}}) # il faut changer le format de la date
#or
db.pubmed_article.aggregate([{$sortByCount:{$year: "$publication_date"}}])                             
```

```python
# https://docs.mongodb.com/manual/reference/operator/aggregation/convert/#exp._S_convert
# https://docs.mongodb.com/manual/reference/operator/aggregation/dateFromString/
# https://docs.mongodb.com/manual/reference/operator/aggregation/toDate/#exp._S_toDate
```

```python
# récupérer le nombre d'articles publiés par an SI FORMAT INCORRECT

// Define stage to add convertedDate field with the converted order_date value

dateConversion = {
   $addFields: {
      convertedDate: { $toDate: "$publication_date" }
   }
};

// Define year by the converted date

defYear = {
   $project: { 
       year:{$year:"$convertedDate"
           }
   }}
};

// Sum year documents 

groupBy = {
    $group:{
        _id:"$year", 
        total:{$sum:1}
    }
};

// Define stage to sort documents by the converted date

sortYear = {
   $sort: { "total": -1 }
};

db.pubmed_article.aggregate( [
   dateConversion,
    defYear,
    groupBy,
   sortYear
])
```
